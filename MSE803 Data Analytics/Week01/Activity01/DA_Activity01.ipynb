```python
import pandas as pd

# Load data
df = pd.read_csv("Data_set_w1A.csv")

print(df)
```

```python
# Find best category in different metrics

best_sales = df.loc[df["sales_sum"].idxmax()]
best_avg = df.loc[df["sales_mean"].idxmax()]
best_quantity = df.loc[df["quantity_sum"].idxmax()]

print("Best Total Sales Category:\n", best_sales)
print("\nBest Average Sales Category:\n", best_avg)
print("\nBest Quantity Sold Category:\n", best_quantity)
```

```python
# Sort by performance
sorted_df = df.sort_values(by="sales_sum", ascending=False)

print("\nSorted by Total Sales:\n", sorted_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Data_set_w1A.csv")

plt.figure()
plt.bar(df["category"], df["sales_sum"])
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales Sum")

plt.show()
```
