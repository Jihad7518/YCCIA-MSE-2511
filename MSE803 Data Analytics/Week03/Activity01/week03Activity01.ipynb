```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("age_networth.csv")

# Display data
print(df.head())
```

```python
# Calculate correlation
correlation = df["Age"].corr(df["Net Worth"])
print("Correlation between Age and Net Worth:", correlation)
```

```python
# Scatter plot
plt.figure()
plt.scatter(df["Age"], df["Net Worth"])
plt.xlabel("Age")
plt.ylabel("Net Worth")
plt.title("Age vs Net Worth")
plt.show()
```

```python
# Regression plot
plt.figure()
sns.regplot(x="Age", y="Net Worth", data=df)
plt.title("Age vs Net Worth with Trend Line")
plt.show()
```

```python
# Heatmap
plt.figure()
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
