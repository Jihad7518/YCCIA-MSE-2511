```python
# Beijing Multi-Site Air Quality - Task 1(Activity-02)

import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load and combine all CSV files
files = glob.glob("PRSA_Data_*.csv")

df_list = [pd.read_csv(file) for file in files]

# Combine all station datasets
df = pd.concat(df_list, ignore_index=True)

print("Dataset Loaded Successfully")
```

```python
# 9. Group Analysis

# By Station
df.groupby('station')['PM2.5'].mean().sort_values(ascending=False)
```

```python
# By Month
df.groupby('month')['PM2.5'].mean()
```

```python
# Monthly Trend
df['datetime'] = pd.to_datetime(df[['year','month','day','hour']])

monthly_pm25 = df.resample('M', on='datetime')['PM2.5'].mean()

monthly_pm25.plot()
plt.title("Monthly PM2.5 Trend")
plt.show()
```

```python
# Outlier Detection
import numpy as np

q1 = df['PM2.5'].quantile(0.25)
q3 = df['PM2.5'].quantile(0.75)
iqr = q3 - q1

outliers = df[(df['PM2.5'] < q1 - 1.5*iqr) | (df['PM2.5'] > q3 + 1.5*iqr)]

print("Number of outliers:", len(outliers))
```

```python
# Station Ranking
station_rank = df.groupby('station')['PM2.5'].mean().sort_values(ascending=False)
station_rank.plot(kind='bar')
plt.title("Average PM2.5 by Station")
plt.show()
```

```python
# Some extra (Advance lil bit)- Linear Regression - Supervised Learning

from sklearn.model_selection import train_test_split

df_ml = df.dropna()

features = ['PM10','SO2','NO2','CO','O3','TEMP','PRES','DEWP']
X = df_ml[features]
y = df_ml['PM2.5']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
print("Model R² Score:", score)
```

```python
# Clustering (K-Means) - Unsupervised
from sklearn.cluster import KMeans

df_cluster = df[['PM2.5','PM10','NO2']].dropna()

kmeans = KMeans(n_clusters=3)
df_cluster['cluster'] = kmeans.fit_predict(df_cluster)

df_cluster.head()
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(max_depth=5)
dt_model.fit(X_train, y_train)

dt_score = dt_model.score(X_test, y_test)
print("Decision Tree R²:", dt_score)
```

```python
print("Linear Regression R²:", score)
print("Decision Tree R²:", dt_score)
```

```python
import pandas as pd

importance = pd.Series(dt_model.feature_importances_, index=X.columns)
importance.sort_values().plot(kind='barh')
plt.title("Feature Importance")
plt.show()
```
