```python
# Beijing Multi-Site Air Quality - Task 1

import pandas as pd
import glob
```

```python
# 1. Load and combine all CSV files

files = glob.glob("PRSA_Data_*.csv")

df_list = []

for file in files:
    temp_df = pd.read_csv(file)
    df_list.append(temp_df)
```

```python
# Combine all station datasets
df = pd.concat(df_list, ignore_index=True)

print("Dataset successfully loaded and combined!\n")
```

```python
# 2. Display first 5 rows

print("First 5 Rows:\n")
print(df.head())
```

```python
# 3. Column names and data types

print("\n🔹 Column Names and Data Types:\n")
print(df.dtypes)
```

```python
# 4. Count total rows and columns

print("\n🔹 Dataset Shape:\n")
print("Total Rows:", df.shape[0])
print("Total Columns:", df.shape[1])
```
