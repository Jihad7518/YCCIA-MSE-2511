```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Create circuit
qc = QuantumCircuit(1, 1)

qc.h(0)
qc.x(0)
qc.measure(0, 0)

# Get simulator
backend = Aer.get_backend("qasm_simulator")

# Run circuit (execute is deprecated)
job = backend.run(qc, shots=1024)
result = job.result()

counts = result.get_counts()
print(counts)

plot_histogram(counts)
plt.show()
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Create circuit: 2 qubits, 2 classical bits
qc2 = QuantumCircuit(2, 2)

# Create Bell state
qc2.h(0)
qc2.cx(0, 1)

# Measure both qubits
qc2.measure([0, 1], [0, 1])

# Visualize circuit
print(qc2.draw())

# Execute
backend = Aer.get_backend('qasm_simulator')
result = execute(qc2, backend, shots=1024).result()
counts = result.get_counts()

# Plot measurement outcomes
plot_histogram(counts)
plt.show()
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from qiskit.visualization import plot_statevector

# Create single-qubit circuit
qc3 = QuantumCircuit(1)

# Apply gates sequentially
qc3.x(0)
qc3.y(0)
qc3.z(0)
qc3.h(0)

# Visualize circuit
print(qc3.draw())

# Get statevector
backend = Aer.get_backend('statevector_simulator')
result = execute(qc3, backend).result()
statevector = result.get_statevector()

# Print final statevector
print("Final Statevector:")
print(statevector)

plot_statevector(statevector)
```

Excercise 04

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_statevector

# Create 3-qubit circuit
qc4 = QuantumCircuit(3)

# Prepare controls in |1>
qc4.x(0)
qc4.x(1)

# Apply Toffoli (CCX)
qc4.ccx(0, 1, 2)

# Visualize circuit
print(qc4.draw())

# Simulate
backend = Aer.get_backend('statevector_simulator')
result = execute(qc4, backend).result()
statevector = result.get_statevector()

print("Final Statevector:")
print(statevector)

plot_statevector(statevector)
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_statevector

# Create 2-qubit circuit
qc5 = QuantumCircuit(2)

# Apply Hadamard on first qubit
qc5.h(0)

# Apply Controlled-Z
qc5.cz(0, 1)

# Visualize circuit
print(qc5.draw())

# Simulate
backend = Aer.get_backend('statevector_simulator')
result = execute(qc5, backend).result()
statevector = result.get_statevector()

print("Statevector showing phase kickback:")
print(statevector)

plot_statevector(statevector)
```
