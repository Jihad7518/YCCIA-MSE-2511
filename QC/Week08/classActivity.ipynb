```python

# Class activity

import qiskit
import requests
import json
import numpy as np

# Use working Quokka server
req_str_qasm = 'http://quokka2.quokkacomputing.com/qsim/qasm'
```

## Quantum Coin Flip
Create superposition using Hadamard gate and measure.

```python
# ===============================
# TASK 1: QUANTUM COIN FLIP
# ===============================

# Create circuit
coin = qiskit.QuantumCircuit(1,1)

# Hadamard
coin.h(0)

# Measurement
coin.measure(0,0)

print(coin)

# ---- Send to Quokka ----
from qiskit.qasm2 import dumps

code = dumps(coin)

data = {
    'script': code,
    'count': 100,
    'state_vector': False
}

result = requests.post(req_str_qasm, json=data)
json_obj = json.loads(result.content)

print("Coin flip results:")
print(json_obj)
```

```python
# 2-BIT QUANTUM RANDOM NUMBER

rand2 = qiskit.QuantumCircuit(2,2)

# Hadamard on both qubits
rand2.h(0)
rand2.h(1)

# Measure
rand2.measure([0,1],[0,1])

print(rand2)

# ---- Send to Quokka ----
from qiskit.qasm2 import dumps

code = dumps(rand2)

data = {
    'script': code,
    'count': 200,
    'state_vector': False
}

result = requests.post(req_str_qasm, json=data)
json_obj = json.loads(result.content)

print("2-bit random results:")
print(json_obj)
```
